Creating my dataframe

```python
import pandas as pd


#creating my dataframe 
charcters = pd.DataFrame({
    'name': ['Gustavo', 'Leon', 'kratos'], # first column names with 3 values strings/object
    'age': [18 , 20 , 25], #second column age with 3 values int64
    'its_hero?': [True, False, True], #thirt column with 3 values bool
    'weight': [80.5, 75.3, 90.6] #fourth column with 3 values float64
})

# charcters.info()
print(charcters)
```

```python
charcters_rename = charcters.rename(columns={
    'name' : 'Full name',  #rename the column but only create a copy
})
print(charcters_rename)
```

```python
charcters.rename(columns={
    'name' : 'Full name'       #after we use inplace true, doesnÂ´t create a copy, change the real name. so when we call the df after this code the column name now is Full name
}, inplace=True)

print(charcters)
```
